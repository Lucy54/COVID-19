```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import pandas_datareader.data as web  
import datetime as dt 
```

North America

```python
# US
us = web.DataReader('IEAXGS', 'fred')
```

```python
fig, ax = plt.subplots(figsize=(10,5))
ax.plot(us.index, us['IEAXGS'], label="U.S.", color='black')

ax.set_ylabel('Millions of Dollars')
ax.set_title('US Exports of Goods & Services')

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.savefig('us_exp.png')

plt.show()
```

South America

```python
# Brazil
brazil = web.DataReader('BRAEXPORTQDSMEI', 'fred')
```

Europe

```python
# NEED TO CONVERT ALL TO SAME CURRENCY
```

```python
# Spain
spain = web.DataReader('ESPEXPORTQDSMEI', 'fred') # in Euros

# UK
uk = web.DataReader('GBREXPORTQDSMEI', 'fred') # in British Pounds

# Norway
nor = web.DataReader('NOREXPORTQDSMEI', 'fred') # in Norwegian Krone
```

```python
fig, ax = plt.subplots(figsize=(10,5))
ax.plot(spain.index, spain['ESPEXPORTQDSMEI']*1.19, label='Spain')
ax.plot(uk.index, uk['GBREXPORTQDSMEI']*1.37, label='UK', color='red')
ax.plot(nor.index, nor['NOREXPORTQDSMEI']*0.12, label='Norway', color='green')

ax.set_ylabel('Dollars')
ax.set_title('Europe: Exports of Goods & Services')

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.savefig('eur_exp.png')

ax.legend()
plt.show()
```
