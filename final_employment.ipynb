```python
# US
# us = web.DataReader('LNS14027662', 'fred')

us = pd.read_excel('us_unemp.xlsx').transpose()
us.drop('Unnamed: 15', axis=0, inplace=True)
us.drop(1, axis=1, inplace=True)
us = us.iloc[1:]
us.index = pd.to_datetime(us.index)
```

```python
unemployment = pd.read_excel('unemployment.xlsx').transpose()
unemployment.drop('Unnamed: 13', axis=0, inplace=True)
unemployment.drop(16, axis=1, inplace=True)
unemployment = unemployment.iloc[1:]
unemployment.columns = ['Dominican Republic', 'Haiti', 'United States', 'Brazil', 'Bolivia', 
                       'Ecuador', 'Spain', 'United Kingdom', 'Norway', 'South Africa', 'Egypt', 
                       'Nigeria', 'China', 'United Arab Emirates', 'Singapore', 'Australia']
unemployment.drop('1990', axis=0, inplace=True)
unemployment.index = pd.to_datetime(unemployment.index)
unemployment
```

```python
# North America
fig, ax = plt.subplots(figsize=(15, 5))

ax.plot(unemployment.index, unemployment['Dominican Republic'], label='Dominican Republic')
ax.plot(unemployment.index, unemployment['Haiti'], label='Haiti')
ax.plot(unemployment.index, unemployment['United States'], label='US')

ax.set_ylabel('Percent')
ax.set_title('North America: Unemployment Rate, % of Total Labor Force')

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.savefig('na_unemp.png')

ax.legend()
plt.show()
```
