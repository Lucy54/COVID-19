```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# national-history (US)
national_hist = pd.read_csv('national-history.csv')

national_hist['date'] = pd.to_datetime(national_hist['date'])

# reverse order of rows : https://stackoverflow.com/questions/35240528/reverse-dataframes-rows-order-with-pandas
national_hist = national_hist.iloc[::-1]
national_hist
```

```python
fig, ax = plt.subplots(figsize=(12,7)) 

ax.plot(national_hist['date'], national_hist['death'])                          

# Despine
sns.despine(ax = ax)                             

# Label 
ax.set_title('COVID-19 Infection Rates in U.S.')
ax.set_ylabel('Number of Deaths')

plt.savefig('us_covid.png')

plt.show()
```

```python
# Historical Data Worldwide (ECDC)

ecdc = pd.read_csv('data.csv', index_col=0)
ecdc.index = pd.to_datetime(ecdc.index)

ecdc = ecdc.iloc[::-1]
ecdc.rename(columns={'countriesAndTerritories':'countries'}, inplace=True)
ecdc.sort_index(inplace=True)
ecdc
```

```python
ecdc = pd.read_csv('data.csv', index_col=0)

# North America
# United States
ecdc_us = ecdc.loc[ecdc['countriesAndTerritories'] == 'United_States_of_America'].copy()
ecdc_us.index = pd.to_datetime(ecdc_us.index)
ecdc_us.sort_index(inplace=True)

# Dominican Republic
ecdc_dr = ecdc.loc[ecdc['countriesAndTerritories'] == 'Dominican_Republic'].copy()
ecdc_dr.index = pd.to_datetime(ecdc_dr.index)
ecdc_dr.sort_index(inplace=True)

# Haiti
ecdc_ht = ecdc.loc[ecdc['countriesAndTerritories'] == 'Haiti'].copy()
ecdc_ht.index = pd.to_datetime(ecdc_ht.index)
ecdc_ht.sort_index(inplace=True)
```

```python
# South America
# Brazil
ecdc_bz = ecdc.loc[ecdc['countriesAndTerritories'] == 'Brazil'].copy()
ecdc_bz.index = pd.to_datetime(ecdc_bz.index)
ecdc_bz.sort_index(inplace=True)

# Bolivia
ecdc_bo = ecdc.loc[ecdc['countriesAndTerritories'] == 'Bolivia'].copy()
ecdc_bo.index = pd.to_datetime(ecdc_bo.index)
ecdc_bo.sort_index(inplace=True)

# Ecuador
ecdc_ec = ecdc.loc[ecdc['countriesAndTerritories'] == 'Ecuador'].copy()
ecdc_ec.index = pd.to_datetime(ecdc_ec.index)
ecdc_ec.sort_index(inplace=True)
```

```python
# Europe
# Spain
ecdc_sp = ecdc.loc[ecdc['countriesAndTerritories'] == 'Spain'].copy()
ecdc_sp.index = pd.to_datetime(ecdc_sp.index)
ecdc_sp.sort_index(inplace=True)

# UK
ecdc_uk = ecdc.loc[ecdc['countriesAndTerritories'] == 'United_Kingdom'].copy()
ecdc_uk.index = pd.to_datetime(ecdc_uk.index)
ecdc_uk.sort_index(inplace=True)

# Norway
ecdc_no = ecdc.loc[ecdc['countriesAndTerritories'] == 'Norway'].copy()
ecdc_no.index = pd.to_datetime(ecdc_no.index)
ecdc_no.sort_index(inplace=True)
```

```python
# Africa
# South Africa
ecdc_sa = ecdc.loc[ecdc['countriesAndTerritories'] == 'South_Africa'].copy()
ecdc_sa.index = pd.to_datetime(ecdc_sa.index)
ecdc_sa.sort_index(inplace=True)

# Egypt
ecdc_eg = ecdc.loc[ecdc['countriesAndTerritories'] == 'Bolivia'].copy()
ecdc_eg.index = pd.to_datetime(ecdc_eg.index)
ecdc_eg.sort_index(inplace=True)

# Nigeria
ecdc_ni = ecdc.loc[ecdc['countriesAndTerritories'] == 'Nigeria'].copy()
ecdc_ni.index = pd.to_datetime(ecdc_ni.index)
ecdc_ni.sort_index(inplace=True)
```

```python
# Asia
# Taiwan
ecdc_tw = ecdc.loc[ecdc['countriesAndTerritories'] == 'Taiwan'].copy()
ecdc_tw.index = pd.to_datetime(ecdc_tw.index)
ecdc_tw.sort_index(inplace=True)

# UAE
ecdc_uae = ecdc.loc[ecdc['countriesAndTerritories'] == 'United_Arab_Emirates'].copy()
ecdc_uae.index = pd.to_datetime(ecdc_uae.index)
ecdc_uae.sort_index(inplace=True)

# Laos
ecdc_ls = ecdc.loc[ecdc['countriesAndTerritories'] == 'Laos'].copy()
ecdc_ls.index = pd.to_datetime(ecdc_ls.index)
ecdc_ls.sort_index(inplace=True)

# China
ecdc_ch = ecdc.loc[ecdc['countriesAndTerritories'] == 'China'].copy()
ecdc_ch.index = pd.to_datetime(ecdc_ch.index)
ecdc_ch.sort_index(inplace=True)

# Singapore
ecdc_sg = ecdc.loc[ecdc['countriesAndTerritories'] == 'Singapore'].copy()
ecdc_sg.index = pd.to_datetime(ecdc_sg.index)
ecdc_sg.sort_index(inplace=True)
```

```python
# Australia
ecdc_aus = ecdc.loc[ecdc['countriesAndTerritories'] == 'Australia'].copy()
ecdc_aus.index = pd.to_datetime(ecdc_aus.index)
ecdc_aus.sort_index(inplace=True)
```

```python
# North America
fig, ax = plt.subplots(figsize=(12,7))
#ax.plot(ecdc_us.index, ecdc_us['deaths'], label='USA')
ax.plot(ecdc_dr.index, ecdc_dr['deaths'], label='Dominican Republic')
ax.plot(ecdc_ht.index, ecdc_ht['deaths'], label='Haiti')

# Despine
sns.despine(ax = ax)                             

# Label 
ax.set_title('COVID-19 Infection Rates in North America')
ax.set_ylabel('Number of Deaths')

plt.savefig('dr-ht.png', bbox_inches="tight")

plt.legend()
plt.show()
```

```python
ecdc_us['growth'] = ecdc_us['deaths'].pct_change() * 100
ecdc_dr['growth'] = ecdc_dr['deaths'].pct_change() * 100
ecdc_ht['growth'] = ecdc_ht['deaths'].pct_change() * 100
```

```python
# North America Percent Change
fig, ax = plt.subplots(figsize=(12,7))
ax.plot(ecdc_us.index, ecdc_us['growth'], label='USA', linestyle='-', marker='o')
ax.plot(ecdc_dr.index, ecdc_dr['growth'], label='Dominican Republic', linestyle='-', marker='o')
ax.plot(ecdc_ht.index, ecdc_ht['growth'], label='Haiti', linestyle='-', marker='o')

# Despine
sns.despine(ax = ax)                             

# Label 
ax.set_title('COVID-19 Infection Growth Rates in North America')
ax.set_ylabel('Percent Change')

plt.savefig('na_covid_growth.png', bbox_inches="tight")

plt.legend()
plt.show()
```
