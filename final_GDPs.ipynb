```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import pandas_datareader.data as web  
import datetime as dt  
import wbdata
import wbgapi as wb
```

```python
gdp = wb.data.DataFrame('NY.GDP.MKTP.CD', ['USA', 'DOM', 'HTI', 'BRA', 'BOL', 'ECU', 
                                         'ESP', 'GBR', 'NOR', 'ZAF', 'EGY', 'NGA', 
                                               'CHN', 'UAE', 'SIN', 'AUS'], 
                             time=range(2004,2020,5), labels=True)
gdp = gdp.drop('Country', axis=1)
gdp.columns = [pd.to_datetime(i[2:]) for i in gdp.columns]
gdp = gdp.transpose()
gdp
```

```python
# DO NOT USE

fig, ax = plt.subplots(figsize=(20,10))

# NA
ax.plot(gdp.index, gdp['DOM'], label='DR', marker='o', linestyle='-')

# SA
ax.plot(gdp.index, gdp['BRA'], label='Brazil', marker='*', linestyle='-')
ax.plot(gdp.index, gdp['BOL'], label='Bolivia', marker='*', linestyle='-')
ax.plot(gdp.index, gdp['ECU'], label='Ecuador', marker='*', linestyle='-')

# Europe
ax.plot(gdp.index, gdp['GBR'], label='United Kingdom', marker='*', linestyle='-')
ax.plot(gdp.index, gdp['ESP'], label='Spain', marker='*', linestyle='-')
ax.plot(gdp.index, gdp['NOR'], label='Norway', marker='*', linestyle='-')

# Africa
ax.plot(gdp.index, gdp['ZAF'], label='South Africa', marker='*', linestyle='-')
ax.plot(gdp.index, gdp['EGY'], label='Egypt', marker='*', linestyle='-')
ax.plot(gdp.index, gdp['NGA'], label='Nigeria', marker='*', linestyle='-')

# Asia
ax.plot(gdp.index, gdp['CHN'], label='China')

# Australia
ax.plot(gdp.index, gdp['AUS'], label='Australia', marker='*', linestyle='-')



ax.set_ylabel('GDP')
ax.set_title('World GDPs')

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

ax.legend()
plt.show()
```
